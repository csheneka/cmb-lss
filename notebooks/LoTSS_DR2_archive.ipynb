```python
def remove_close_neighbors(distances, min_distance):
    # Pick distances which are small enough
    small_distances = distances.loc[distances['distance'] < min_distance].reset_index(drop=True)
    distances_reduced = distances.loc[distances['distance'] >= min_distance].reset_index(drop=True)
    
    print(small_distances.shape)
    print(distances_reduced.shape)
    
    # Create groups of removed and kept objects
    id_to_remove = []
    id_to_keep = []
    for i in tqdm(range(small_distances.shape[0])):
        # Remove if any object ID in pair already seen before
        to_remove = (
            small_distances.loc[i, 'ID_a'] in id_to_remove
            or small_distances.loc[i, 'ID_a'] in id_to_keep
            or small_distances.loc[i, 'ID_b'] in id_to_remove
            or small_distances.loc[i, 'ID_b'] in id_to_keep
        )
        
        # If both IDs not seen, then add one to keep, remove second
        if to_remove:
            id_to_remove.append(small_distances.loc[i, 'ID_a'])
        else:
            id_to_keep.append(small_distances.loc[i, 'ID_a'])
        id_to_remove.append(small_distances.loc[i, 'ID_b'])
        
    print(len(id_to_keep))
    print(len(id_to_remove))

    # Remove objects to remove from the remaining distances
    idx_reduced = (~ distances_reduced['ID_a'].isin(id_to_remove)) & (~ distances_reduced['ID_b'].isin(id_to_remove))
    distances_reduced = distances_reduced.loc[idx_reduced].reset_index(drop=True)
    
    print(distances_reduced.shape)
    
    return distances_reduced
```

```python
# TODO: show distance in arcmin
def plot_multiple_log_hist(data_arr, n_bins=100, x_label=None, y_label=None):
    bins = np.logspace(np.log10(np.min(data_arr[0][0])), np.log10(np.max(data_arr[0][0])), n_bins)
    for data, data_name in data_arr:
        plt.hist(data, bins=bins, histtype='step', label=data_name)
    plt.gca().set_xscale('log')
    plt.gca().set_yscale('log')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(loc='upper left')
    plt.show()
```
